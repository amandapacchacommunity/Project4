# Advancement Portfolio and Donor Retention Analysis

## Data Validation and Preparation

This notebook generates synthetic nonprofit donor datasets and performs initial validation and exploratory analysis.

### Key Steps

1. Load analysis libraries
2. Generate synthetic donor, membership, and pledge datasets
3. Validate dataset structure
4. Inspect sample records
5. Visualize membership distribution and pledge status


```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns

print("Libraries loaded successfully")
```

```python
# Generate synthetic nonprofit datasets

np.random.seed(42)

n_donors = 500

donors = pd.DataFrame({
    "donor_id": range(1, n_donors + 1),
    "donor_type": np.random.choice(["individual", "foundation", "corporate"], n_donors, p=[0.85, 0.10, 0.05]),
    "engagement_score": np.random.randint(1, 100, n_donors),
    "lifetime_giving": np.random.gamma(2, 2000, n_donors)
})

memberships = pd.DataFrame({
    "donor_id": np.random.choice(donors["donor_id"], 300),
    "year": np.random.choice([2022, 2023, 2024], 300),
    "membership_level": np.random.choice(["basic", "supporter", "patron"], 300),
    "amount": np.random.choice([50, 100, 250, 500], 300)
})

pledges = pd.DataFrame({
    "donor_id": np.random.choice(donors["donor_id"], 200),
    "pledge_amount": np.random.randint(500, 10000, 200),
    "year": np.random.choice([2022, 2023, 2024], 200),
    "status": np.random.choice(["paid", "outstanding"], 200, p=[0.7, 0.3])
})

print("Synthetic datasets generated successfully")
```

```python
# Dataset overview

print("Donor dataset size:", donors.shape)
print("Membership records:", memberships.shape)
print("Pledge records:", pledges.shape)

display(donors.head())
display(memberships.head())
display(pledges.head())
```

```python
# Membership distribution

memberships["membership_level"].value_counts().plot(kind="bar")

plt.title("Membership Level Distribution")
plt.xlabel("Membership Level")
plt.ylabel("Count")

plt.show()
```

```python
# Pledge status distribution

pledges["status"].value_counts().plot(kind="bar")

plt.title("Pledge Status Distribution")
plt.xlabel("Status")
plt.ylabel("Number of Pledges")

plt.show()
```
